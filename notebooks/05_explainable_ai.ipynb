```python
import pandas as pd
import shap
import joblib
import matplotlib.pyplot as plt

print("Libraries Loaded Successfully")
```

```python
df = pd.read_csv("../data/predictive_maintenance.csv")

print(df.shape)

df.head()
```

```python
df = df.drop(
    ["UDI", "Product ID"],
    axis=1
)

df = pd.get_dummies(
    df,
    columns=["Type"],
    drop_first=True
)

df.columns = (
    df.columns
      .str.replace('[', '', regex=False)
      .str.replace(']', '', regex=False)
      .str.replace(' ', '_')
)

print(df.columns)
```

```python
df["Temp_Diff"] = (
    df["Process_temperature_K"]
    - df["Air_temperature_K"]
)

df["Power"] = (
    df["Torque_Nm"]
    * df["Rotational_speed_rpm"]
)

df.head()
```

```python
X = df.drop(
    "Machine_failure",
    axis=1
)

print(X.shape)
```

```python
model = joblib.load(
    "../models/xgboost_model_v2.pkl"
)

print(type(model))
```

```python
explainer = shap.TreeExplainer(model)

print("Explainer Created")
```
